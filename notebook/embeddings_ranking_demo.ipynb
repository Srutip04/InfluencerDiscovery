```python
import numpy as np, pandas as pd
creators = pd.DataFrame({
 'creator_id': range(1,11),
 'bio': [
 'Fashion & lifestyle influencer','Tech reviewer','Fitness coach','Beauty tutorials','Travel photographer',
 'Gaming streamer','Food blogger','Parenting hacks','Comedy skits','Business entrepreneurship'
 ],
 'followers': np.random.randint(10000,500000,10),
 'engagement': np.random.uniform(0.5,6.0,10)
})
creators
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(creators['bio'])
embeddings = normalize(X.toarray())
embeddings[:3]
```

```python
query='fashion beauty influencer'
q=normalize(vectorizer.transform([query]).toarray())[0]
creators['sim_score']=(embeddings@q)
creators.sort_values('sim_score',ascending=False)
```

```python
creators['log_followers']=np.log1p(creators['followers'])/np.log(500000)
creators['rank_score']=creators['sim_score']*0.6 + creators['engagement']/6*0.25 + creators['log_followers']*0.15
creators.sort_values('rank_score',ascending=False)
```
